# Gaborov filter

U obradi slika, Gaborov filter je <b>linearni filter</b> koji se koristi za analizu teksture. 
Pomocu Gaborovog filtera slika se prostorno i frekvencijski lokalizuje tako da ulazni uzorak bude sto je moguce bolje vremensko - frekvencijski odredjen. Zbog toga Gaborov filter ima sirok spektar primene.<br> 
Pomocu razlicitih Gaborovih filtera mozemo izvrsiti ekstrakciju zanimljivih svojstava iz slike.<br>
Na primer, mozemo analizirati da li na slici postoji sadrzaj odredjene frekvencije u zeljenom smeru.<br>Neke od pimena su:
* Detekcija ivica objekta na slici
* Detekcija lica na slici
* Detekcija tesko uocljivih objekata na slici
* Ulepsavanje slike - za pogodno odabrane parametre filtera mozemo dobiti vrlo interesantne filtracije slika
* Formiranje banki filtera za eksperimente u oblasti masinskog ucenja
* U Medicini - Klasifikacija Alchajmerove bolesti na osnovu slike
* U Medicini - Detekcija glave ocnog zivca kod pacijenata sa glaukomom

## Gaborova funkcija

Da bismo videli kako funkcionise Gaborov filter pozabavimo se prvo Gaborovom funkcijom.<br>
Gaborova funkcija je kompleksno vrednosna funkcija koja kao ulaz pored mnogih parametara zahteva i (x, y) koordinatu tacke 2D teksture za koju se racuna vrednost Gaborove funkcije.<br>
Gaborov filter nije nista drugo nego <b>Gausova kernel funkcija</b> modulirana pomocu <b>kompleksne sinusoide</b>.<br>
Gausova 2D funkcija ima oblik: w(x, y) = $Ae^{-({\frac{(x-x_0)^2} {2\sigma_x^2}} + {\frac{(y-y_0)^2} {2\sigma_y^2}})}$ <br>
Gde je:
* A - amplituda Gausove funkcije
* Tacka $({x_0}, {y_0})$ - centar Gausove funkcije
* ${\sigma_x}, {\sigma_y}$ - standardne devijacije po x i po y osi, definisu 'debljinu' Gausove funkcije

<br>
Primetimo da je Gausova funkcija realno - vrednosna funkcija.

Kompleksna sinusoida ima oblik: s(x, y) = $e^{i(2\pi({u_0x + v_0y}) + {\phi})}$.<br>
Gde su ${u_0}$ i ${v_0}$ redom prostorne frekvencije sinusoide po x i po y.<br> 

Zapisimo sada ${u_0}$ i ${v_0}$ u polarnim koordinatama uzimajuci: <br>
* $f$ = ${\sqrt{u_0^2 + v_0^2}}$ <br>
* ${u_0}$ = $f\cos{\theta}$<br>
* ${v_0}$ = $f\sin{\theta}$<br>

Konacno, nasa kompleksna sinusoide ima oblik: s(x,y) = $e^{i(2\pi f(xcos\theta + ysin\theta) + \psi)}$

Gaborova funkcija tada ima oblik: <br>
${g(x,y,f,\theta, \psi, \sigma_x, \sigma_y)} = Ae^{-({\frac{(x-x_0)^2} {2\sigma_x^2}} + {\frac{(y-y_0)^2} {2\sigma_y^2}}) }* e^{i(2\pi f(xcos\theta + ysin\theta) + \psi)}$<br>

Ovako smo dobili jedan vrlo opst oblik Gaborove funkcije. <br>
Skaliranjem parametara i uvodjenjem novih promenljivih dobija se sledeci oblik Gaborove funkcije:<br>
${g(x,y;\lambda, \theta, \psi, \sigma, \gamma)} = e^{-{\frac{x'^2 + \gamma'^2y'^2} {2\sigma^2}} }* e^{i(2\pi\frac {x'}{\lambda} + \psi )}$, gde je:<br>
* $x' = xcos\theta - ysin\theta$<br> 
* $y' = xsin\theta + ycos\theta$<br>
* $\lambda$ - talasna duzina Gaborove funkcije
* $\theta$ - ugao koji kontrolise orijentaciju Gaborove funkcije
* $\psi$ - fazni pomeraj Gaborove funkcije
* $\sigma$ - standardna devijacija Gausove funkcije po x osi
* $\gamma$ - parametar koji pravi razliku izmedju standardne devijacije Gausove funkcije po x i po y osi. Vazi pravilo : $\sigma = \sigma_x = \gamma\sigma_y $

Pomocu ovakvih parametara mozemo u potpunosti kontrolisati izgled Gaborovog filtera. Ako zelimo da generisemo Gaborov filter kao sliku dimenzija N x M, onda je portrebno zamisliti je u nekom koordinatnom sistemu. (x, y) mozemo shvatiti kao koordinate piksela u tom koordinatnom sistemu ciju vrednost zelimo da odredimo. Generisacemo nekoliko filtera gde cemo prikazati kako gore opisani parametri uticu na izgled Gaborovog filtera.<br>

U nasim eksperimentima koristicemo realni deo Gaborove funkcije.<br>
Kada generisemo Gaborov filter mozemo izvrsiti konvoluciju filtera sa slikom (predstavljenom u matricnom obliku) radi ekstrakcije odredjenih svojstava slike.

## Implementacija Gaborove funkcije i izracunavanje Gaborovog filtera

```python
import math
import numpy as np
from matplotlib import pyplot as plt
import cv2
from PIL import Image
```

```python
def gabor_function(x, y, Lambda, theta, psi, sigma, gamma):
    
    sigma_x = sigma
    sigma_y = sigma / gamma
    x_theta = x * math.cos(theta) + y * math.sin(theta)
    y_theta = -x * math.sin(theta) + y * math.cos(theta)
    
    return math.exp(-.5 * (x_theta**2 / sigma_x**2 + y_theta**2 / sigma_y**2)) * math.cos(2*math.pi / Lambda * x_theta + psi) 
```

```python
def gabor_filter(size_x, size_y, Lambda, theta, psi, sigma, gamma):
    
    M = np.zeros((size_x, size_y))
    x = -size_x / 2
    
    for i in range(size_x):
        y = -size_y / 2
        for j in range(size_y):
            M[i, j] = gabor_function(x, y, Lambda, theta, psi, sigma, gamma)
            y = y + 1
        x = x + 1
        
    return M
```

Da bismo lepo videli kako parametri Gaborove funkcije uticu na izgled samog Gaborovog filtera prikazacemo po 4 filtera dimenzije 50 x 50 gde cemo jedan parametar varirati a ostale fiksirati.<br>
Prvo variramo parametar $\theta$ (theta) koji odredjuje orijentaciju filtera:

```python
sigma = 20
lam = 50 
gamma = 0.4
N = 4

for i in range(1, N, 2):
    plt.subplot(1, 2, 1, title = "theta = {index}pi / {parts}".format(index = i, parts =  N))
    plt.imshow(gabor_filter(50, 50, lam, math.pi * i / N, 0, sigma, gamma))
    plt.subplot(1, 2, 2, title = "theta = {index}pi / {parts}".format(index = i + 1, parts =  N))
    plt.imshow(gabor_filter(50, 50, lam, math.pi * (i+1) / N, 0, sigma, gamma))
    plt.show()
```

Variramo parametar $\lambda$ - talasnu duzinu Gaborove funkcije: 

```python
for i in range(1, N, 2):
    plt.subplot(1, 2, 1, title = "lambda = {lam}".format(lam = lam * i / N))
    plt.imshow(gabor_filter(50, 50, 2 * lam * i / N, 0, 0, sigma, gamma))
    plt.subplot(1, 2, 2, title = "lambda = {lam}".format(lam = lam * (i + 1) / N))
    plt.imshow(gabor_filter(50, 50, 2* lam * (i+1) / N, 0, 0, sigma, gamma))
    plt.show()
```

 Variramo parametar $\sigma$ - standardnu devijaciju po x osi:

```python
for i in range(1, N, 2):
    plt.subplot(1, 2, 1, title = "sigma = {sigma}".format(sigma = sigma * i / N))
    plt.imshow(gabor_filter(50, 50, lam, 0, 0, sigma * i / N, gamma))
    plt.subplot(1, 2, 2, title = "sigma = {sigma}".format(sigma = sigma * (i + 1) / N))
    plt.imshow(gabor_filter(50, 50, lam, 0, 0, sigma * (i + 1) / N, gamma))
    plt.show()
```

Varirajmo $\gamma$ parametar koji pravi razliku izmedju standardne devijacije po x i po y:

```python
gamma = 2.4

for i in range(1, N, 2):
    plt.subplot(1, 2, 1, title = "gamma = {gamma}".format(gamma = round(gamma * i / N, 2)))
    plt.imshow(gabor_filter(50, 50, lam, 0, 0, sigma, round(gamma * i / N, 2)))
    plt.subplot(1, 2, 2, title = "gamma = {gamma}".format(gamma = round(gamma * (i + 1) / N, 2)))
    plt.imshow(gabor_filter(50, 50, lam, 0, 0, sigma, gamma * (i + 1) / N))
    plt.show()
```

Varirajmo parametar $\psi$ - fazni pomeraj Gaborove funkcije. Jako cesto se ovaj parametar fiksira na 0.

```python
psi = 120
for i in range(0, N, 2):
    plt.subplot(1, 2, 1, title = "psi = {psi}".format(psi = psi * i / N))
    plt.imshow(gabor_filter(50, 50, lam, 0,  psi * i / N, sigma, gamma))
    plt.subplot(1, 2, 2, title = "psi = {psi}".format(psi = psi * (i + 1) / N))
    plt.imshow(gabor_filter(50, 50, lam, 0, psi * (i+1) / N, sigma, gamma))
    plt.show()
```

## Primer 1 : Detekcija uljeza na slici 

Za prvi primer upotrebe Gaborovog filtera ucitacemo sliku na kojoj se uljez.<br>
Posto se uljez krije iza periodicne teksture koja se sastoji od horizontalnih linija/traka, mi cemo pokusati da ga detektujemo koriscenjem vertikalnog Gaborovog filtera.  

Ucitajmo sliku sa uljezom:

```python

intruder = cv2.imread('data/hidden.png')
intruder = cv2.cvtColor(intruder, cv2.COLOR_BGR2GRAY)
plt.imshow(intruder)
plt.show()
```

Pripremimo filter za detekciju uljeza:

```python
intruder_filter = gabor_filter(30, 30, Lambda = 15, theta = math.pi/2, psi = 0, sigma = 7, gamma = 10)
plt.imshow(intruder_filter)
plt.show()
```

Konacno, uljeza cemo detektovati kao kontrast u odnosu na pozadinu na slici koja sledi.

```python
intruder_filtered = cv2.filter2D(intruder, cv2.CV_8UC3, intruder_filter)
plt.imshow(intruder_filtered)
plt.show()
```

Sacuvajmo rezultat naseg eksperimenta:

```python
Image.fromarray(intruder_filtered).save('./output/intruder.png')
```

## Primer 2:  Otklanjanje smetnji sa slike

Za drugi primer upotrebe Gaborovog filtera ucitacemo sliku misteriozne zivotinje. <br>Slika je puna suma i zbog toga je tesko utvrditi o kojoj zivotinji je rec.<br>
Nas cilj je da otklonimo sum.<br><br>
Ucitajmo sliku i prikazimo je:

```python
hidden_cat = cv2.imread('data/cat.png')
hidden_cat = cv2.cvtColor(hidden_cat, cv2.COLOR_BGR2GRAY)
plt.imshow(hidden_cat, cmap = 'gray')
plt.show()
```

Pripremimo filter za otklanjanje suma.<br>
Insistiracemo da filter bude orijentisan horizontalno, radi eliminacije vertikalnih smetnji.<br>
Uz pogodno odabrane parametre dobijamo filter:

```python
cat_filter = gabor_filter(15, 15, Lambda = 5, theta = 0, psi = 0, sigma = 7, gamma = 0.2*0.9)
plt.imshow(cat_filter)
plt.show()
```

```python
filtered_cat = cv2.filter2D(hidden_cat, cv2.CV_8UC3, cat_filter)
```

Nakon filtracije dobijamo dosta jasniju sliku:

```python
plt.imshow(filtered_cat, cmap = 'gray')
plt.show()
```

Nije tesko utvrditi da je misteriozna zivotinja macka.

```python
Image.fromarray(filtered_cat).save('./output/cat.png')
```

## Primer 3: Izdvajanje srafura i pravaca  

U ovom primeru ucitacemo sliku zebre i sa nje cemo izdvojiti samo pravce odredjenog nagiba.

```python
zebra  = cv2.imread('data/zebra.jpg')
plt.imshow(zebra)
plt.show()
```

Izdvojicemo pravce sa nagibom od : $\frac{\pi}{4}, \frac{\pi}{6}, \frac{4\pi}{5}, \pi$

```python
pi = math.pi
angles = {'pi/4': pi / 4, 'pi/6': pi / 6, '4pi/5': 4 * pi / 5, 'pi': pi}
counter = 1

for angle in angles:
    
    zebra_filter = gabor_filter(100, 100, Lambda = 60, theta = angles[angle], psi = 0, sigma = 20, gamma = 0.4)
    zebra_filtered = cv2.filter2D(zebra, cv2.CV_8UC3, zebra_filter)
    
    plt.imshow(zebra_filtered)
    plt.title(angle)
    plt.show()
    
    Image.fromarray(zebra_filtered).save('./output/zebra{counter}.png'.format(counter = counter))
    counter = counter + 1
```
